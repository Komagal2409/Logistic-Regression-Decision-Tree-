```python
# decision tree #gini impurtiy #entropy and high information
import pandas as pd
df = pd.read_csv("salaries.csv")
df
```

```python
df.head()
```

```python
inputs = df.drop('salary_more_then_100k',axis='columns')
target = df['salary_more_then_100k']
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
inputs.head()
```

```python
final = inputs.drop(['company', 'job', 'degree'],axis='columns')
final
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(final,target)
```

```python
model.score(final,target)
```

```python
model.predict([[2,2,1]])
```

```python
model.predict([[1,1,0]])
```

```python
import pandas as pd
d = pd.read_csv("titanic.csv")
d
```

```python
clf = d.drop([ 'Survived', 'PassengerId', 'Name', 'SibSp', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns')
targett = d['Survived']
clf
```

```python
targett
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex = LabelEncoder()
clf['Sex_n'] = le_Sex.fit_transform(clf['Sex'])
clf
```

```python
finall = clf.drop(['Sex'],axis='columns')
finall
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(finall,targett)
```

```python
model.score(finall,targett)
```

```python
model.predict([[3,22.0,7.25,1]])
```
